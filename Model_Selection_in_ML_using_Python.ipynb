splitting features and target

```python
X=data.drop(columns='target', axis=1) #splitting dataset into features by dropping target
y=data['target'] #selecting target column as target feature
print(X)
print(y)
```

```python
# converting the features (X) and target variable (y) to numpy arrays
# this is done for better compatibility with machine learning libraries and for performance reasons
X = np.asarray(X)
y = np.asarray(y)
```

model selection

```python
# importing the models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#list of models
models = [LogisticRegression(max_iter=10000), SVC(), KNeighborsClassifier(), RandomForestClassifier(random_state=0)]
```

```python
#creating dictionary that contains hyperparameter values for the models
model_hyperparameters={
    'log_reg_hypara':{
        'C':[1,5,10,20]
    },
    'svc_hyperparameters':{
        'kernel':['linear','poly','rbf','sigmoid'],
        'C':[1,5,10,20]
    },

    'KNN_hyperparameters':{
        'n_neighbors':[3,5,10]

    },

    'random_forest_hyperparameters':{
        'n_estimators':[10,20,50,100]
    }
}
```

```python
#creating listof key values
model_keys = list(model_hyperparameters.keys())
print(model_keys)
```

applying GridSearchCV

```python
#importing necessary modules for model evaluation and hyperparameter tuning
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
#this function performs model selection and hyperparameter tuning using GridSearchCV
def ModelSelection(models,model_hyperparameters):
  result = []
  i=0
  for model in models:
    key=model_keys[i]
    params=model_hyperparameters[key]
    i+=1
    print(model)
    print(params)

    classifier=GridSearchCV(model,params,cv=5)
    #fitting data into classifier
    classifier.fit(X,y)
    result.append({
        'model_used':model,
        'highest_score':classifier.best_score_,
        'best_hyperparameters':classifier.best_params_
    })
  result_dataframe=pd.DataFrame(result,columns=['model_used','highest_score','best_hyperparameters'])
  return result_dataframe
```

```python
ModelSelection(models,model_hyperparameters)
```
